```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
```

```python
data = np.hstack([np.array([12,4,-2,7,3,-9,10,2,2,1,-3,7,25,-9,2,0,-18,-3,-4]),
                 np.random.randint(-15,20, (10,))])
```

```python
data.shape
```

```python
s = pd.Series(data)
```

```python
fig, axes = plt.subplots(2,1, figsize=(10,5), gridspec_kw={'height_ratios': [8, 1]})
plt.axis('off')
sns.lineplot(x=range(data.shape[0]), y=data, ax=axes[0])
axes[0].set_xlim(0, len(data)-1)
axes[0].set_xticks([])
axes[0].set_yticks(np.arange(-20, 30, 5))
sns.heatmap(data[None,:], ax=axes[1], annot=True, fmt="d", xticklabels=False, yticklabels=False, cbar=None)
```
